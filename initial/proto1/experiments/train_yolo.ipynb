```python
# ✅ 1. Setup
from ultralytics import YOLO
import os
import matplotlib.pyplot as plt
from IPython.display import Image, display
import glob

# Optional: Check if GPU is available
!nvidia-smi

# ✅ 2. Define dataset and model path
dataset_yaml = r"D:\target\AI-AGENTS\accident_detection_ai\datasets\labeled\data.yaml"

model_name = "yolov8n.pt"  # You can change to 'yolov8s.pt', 'yolov8m.pt', etc.

# ✅ 3. Load the YOLO model
model = YOLO(model_name)  # Load pretrained YOLOv8 model

# ✅ 4. Train the model
results = model.train(
    data=dataset_yaml,
    epochs=50,
    imgsz=640,
    batch=16,
    name="accident_yolov8n",
    project="experiments/runs",  # output directory
    save=True,
    save_period=5,  # save model every 5 epochs
    patience=10,
    val=True
)

```

```python
# ─────────────────────────────────────────────────────────────────────────────
# 1. Install any missing dependencies (run once)
# ─────────────────────────────────────────────────────────────────────────────
# If you haven’t already installed ultralytics and onnx, uncomment these lines:
# !pip install ultralytics --quiet
# !pip install onnx --quiet

# ─────────────────────────────────────────────────────────────────────────────
# 2. Imports
# ─────────────────────────────────────────────────────────────────────────────
from ultralytics import YOLO
from IPython.display import Image, display
import glob
import os

# ─────────────────────────────────────────────────────────────────────────────
# 3. Display Training Curves
# ─────────────────────────────────────────────────────────────────────────────
# Adjust this path to match exactly where 'results.png' was saved during training:
training_results_path = r"D:\target\AI-AGENTS\accident_detection_ai\experiments\experiments\runs\accident_yolov8n2\results.png"
display(Image(filename=training_results_path))

# ─────────────────────────────────────────────────────────────────────────────
# 4. Load the Best Model
# ─────────────────────────────────────────────────────────────────────────────
best_model_path = r"D:\target\AI-AGENTS\accident_detection_ai\experiments\experiments\runs\accident_yolov8n2\weights\best.pt"
best_model = YOLO(best_model_path)

# ─────────────────────────────────────────────────────────────────────────────
# 5. Run Inference on Test Set
# ─────────────────────────────────────────────────────────────────────────────
# This will read all images under datasets/labeled/test/images,
# save predictions (images + TXT) under runs/detect/predict/
test_images_dir = r"D:\target\AI-AGENTS\accident_detection_ai\datasets\labeled\test\images"
results = best_model.predict(
    source=test_images_dir,
    conf=0.25,
    save=True,     # write out predicted images to runs/detect/predict/
    save_txt=True  # write out predicted bounding boxes in .txt format
)

# Display the first 5 prediction results
predicted_images = glob.glob("runs/detect/predict/*.jpg")
for img_path in predicted_images[:5]:
    display(Image(filename=img_path))

# ─────────────────────────────────────────────────────────────────────────────
# 6. Evaluate on Validation/Test Set
# ─────────────────────────────────────────────────────────────────────────────
# Pass your data.yaml so YOLO knows where the validation images and labels live
data_yaml = r"D:\target\AI-AGENTS\accident_detection_ai\datasets\labeled\data.yaml"
metrics = best_model.val(data=data_yaml)
print("Evaluation metrics:\n", metrics)

# ─────────────────────────────────────────────────────────────────────────────
# 7. Export to ONNX (now that onnx is installed above)
# ─────────────────────────────────────────────────────────────────────────────
exported_model_path = best_model.export(format="onnx")  # exports to onnx and returns the file path
print(f"✅ Model exported to: {exported_model_path}")
```
